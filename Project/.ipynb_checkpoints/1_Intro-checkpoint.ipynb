```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
ax = df["job_country"].value_counts().head(10).sort_values().plot(kind="barh", xlabel="Job Postings (Q)", ylabel="")
ax.bar_label(ax.containers[0], padding=5)
plt.xlim(0,250_000)
plt.title("Top 10 Countries with Most Job Postings")
plt.show()
```
